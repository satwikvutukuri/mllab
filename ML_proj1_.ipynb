# **binomial distribution**

```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv('https://raw.githubusercontent.com/satwikvutukuri/mllab/main/data1.csv')
for col in ['wise Malaria cases in study area', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 10']:
    data = df[col].dropna()
    n_binomial = len(data)
    p_binomial = data.mean() / data.max()
    log_likelihood_binomial = np.sum(stats.binom.logpmf(data, n_binomial, p_binomial))
    k_binomial = 2 # Number of parameters in Binomial
    aic_binomial = -2 * log_likelihood_binomial + 2 * k_binomial
    results[col]['Binomial'] = {'AIC': aic_binomial}
    print(f"Binomial distribution for year {col}: AIC = {aic_binomial}")
```

# **Discrete Uniform distribution**

```python
for col in ['wise Malaria cases in study area', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 10']:
    data = df[col].dropna()
    # Assuming the minimum and maximum values for the discrete uniform distribution
    low = data.min()
    high = data.max()
    # Calculate log-likelihood (assuming equal probability for each value)
    log_likelihood_uniform = np.sum(np.log(1/(high - low + 1)) * np.ones_like(data))
    k_uniform = 2  # Number of parameters in Discrete Uniform
    aic_uniform = -2 * log_likelihood_uniform + 2 * k_uniform
    results[col]['Discrete Uniform'] = {'AIC': aic_uniform}
    print(f"Discrete Uniform distribution for year {col}: AIC = {aic_uniform}")
```

# **Poisson Binomial distribution**
