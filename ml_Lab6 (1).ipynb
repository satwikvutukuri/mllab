```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import load_iris
```

# Load the iris dataset, which is a built-in dataset from sklearn

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

# Select the first two features (sepal length and sepal width) from the dataset to simplify the visualization

```python
X = iris.data[:, :2]
y = iris.target
```

# Initialize SVM classifiers with different kernel types and parameters

```python
svc_linear = svm.SVC(kernel='linear', C=1.0)
svc_poly = svm.SVC(kernel='poly', degree=3, C=1.0)
svc_rbf = svm.SVC(kernel='rbf', gamma=1, C=1.0)
svc_sigmoid = svm.SVC(kernel='sigmoid', gamma=1, C=1.0)

svc_linear.fit(X, y)
svc_poly.fit(X, y)
svc_rbf.fit(X, y)
svc_sigmoid.fit(X, y)
```

# Function to plot the decision boundaries for a given model

```python
def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    h = 0.02
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
```

# Create a figure for plotting 4 subplots

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plot_decision_boundary(svc_linear, X, y)
plt.title('Linear Kernel')

plt.subplot(1, 4, 2)
plot_decision_boundary(svc_poly, X, y)
plt.title('Polynomial Kernel (degree 3)')

plt.subplot(1, 4, 3)
plot_decision_boundary(svc_rbf, X, y)
plt.title('RBF Kernel')

plt.subplot(1, 4, 4)
plot_decision_boundary(svc_sigmoid, X, y)
plt.title('Sigmoid Kernel')
```

# Display the plot

```python
plt.show()
```
